```python
import pdfplumber
import pandas as pd
import numpy as np
```

```python
# Provide the path to the PDF file
file_path = r"D:\OneDrive - Nexus365\[3] Projects\JUPAS Calculator\universities\cuhk\useful_information_2023.pdf"
```

```python
# Extract the text from the PDF file into a table

tables = []

with pdfplumber.open(file_path) as pdf:
    for page in pdf.pages:
        table = page.extract_table()
        tables.append(table)

```

```python
# Combine the tables into one
df = pd.DataFrame()
for table in tables:
    df = pd.concat([df, pd.DataFrame(table)], ignore_index=True)

df = df.fillna('')
# Display max rows = 100
pd.set_option('display.max_rows', 452)

# Remove rows that are completely empty
df = df[~df.eq('').all(1)]

# Export the table to an excel file
df.to_excel(r"D:\OneDrive - Nexus365\[3] Projects\JUPAS Calculator\universities\cuhk\cuhk.xlsx", index=False)

# MANY ERRORS


```
